```python
import os, json

dfs = []
json_files = [
    pos_json for pos_json in os.listdir("./output") if pos_json.endswith(".json")
]
for file in json_files:
    data = pd.read_json(f"./output/{file}")
    dfs.append(data)
df = pd.concat(dfs, ignore_index=True)
df
```
